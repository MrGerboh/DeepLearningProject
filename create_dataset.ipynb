```python
from datasets import load_dataset
import py7zr
import os

# Load the IMDB reviews dataset from Hugging Face
dataset = load_dataset("imdb")

# Create a directory to save files if it doesn't exist
output_dir = "compressed_texts"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate through the dataset and save each text along with its compressed version
for idx, example in enumerate(dataset['train']):
    text = example['text']

    # Save the original text
    with open(os.path.join(output_dir, f"original_text_{idx}.txt"), "w") as f:
        f.write(text)

    # Compress the text using 7-zip
    with py7zr.SevenZipFile(os.path.join(output_dir, f"compressed_text_{idx}.7z"), 'w') as archive:
        archive.write(os.path.join(output_dir, f"original_text_{idx}.txt"), f"original_text_{idx}.txt")

    print(f"Text {idx} compressed and saved.")

print("All texts compressed and saved.")
```
