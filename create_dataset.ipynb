```python
from datasets import load_dataset
import pylzma
import os

# Load only the train and test datasets from IMDB
dataset = load_dataset("imdb")

dataset.pop('unsupervised')

# Function to compress text using 7-zip
def compress_text(text):
    compressed_data = pylzma.compress(text.encode())
    return compressed_data

# Add a new column to the dataset with compressed text
dataset = dataset.map(lambda example: {"text_compressed": compress_text(example["text"])}, remove_columns=["label"])

# Set the path to save the new dataset
output_path = "compressed_imdb_dataset"

# Save the dataset to disk
dataset.save_to_disk(output_path)

# Verify that the compressed dataset is saved
print(f"Dataset saved to: {output_path}")

# Optionally, you can load and check the saved dataset
loaded_dataset = load_dataset(output_path)
print(loaded_dataset["train"][0])
```
